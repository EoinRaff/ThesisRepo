```python
def slope(x, a):
    return (x**a)/((x**a) + ((1-x)**a))
```

```python
import numpy as np

x_axis = np.linspace(0, 1, 11)
print(x_axis)
```

```python
original_slope = [slope(x, -3.5)*100 for x in x_axis ]
print((original_slope))
```

```python
import matplotlib.pyplot as plt

plt.plot(x_axis, original_slope)
```

```python
def landslide(sample, sensitivity, strength):
    for i in range (1, len(sample)-1):
        for j in range(-1, 2):
            if sample[i] > sample[i+j] + sensitivity:
                sample[i+j] = sample[i+j] + strength
                sample[i] = sample[i] - strength
    return sample
```

```python
erosion_1 = landslide(np.copy(original_slope), 20, 10)
erosion_2 = landslide(np.copy(original_slope), 10, 5)
erosion_3 = landslide(np.copy(original_slope), 5, 2)
```

```python

plt.figure()
plt.plot(x_axis, original_slope)
plt.plot(x_axis, erosion_1)
plt.legend(["Before Erosion", "After Erosion"])

plt.figure()
plt.plot(x_axis, original_slope)
plt.plot(x_axis, erosion_2)
plt.legend(["Before Erosion", "After Erosion"])


plt.figure()
plt.plot(x_axis, original_slope)
plt.plot(x_axis, erosion_3)
plt.legend(["Before Erosion", "After Erosion"])

```
